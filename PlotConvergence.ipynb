```python
import numpy as np
import matplotlib.pyplot as plt

fpath = "/pscratch/sd/b/bthorne/fairuniverse/hsc_dataset/kappa_maps/0_realization0_Nmesh8192_plane43_empty10_convergence_coherent.npy"
convergence = np.load(fpath)
```

```python
pixel_extent = [0, 8192, 0, 8192]

angular_extent = [0, 16, 0, 16]  # in degrees

# Linear mapping from pixels to degrees
x_ticks_pixel = np.linspace(pixel_extent[0], pixel_extent[1], 9)  # 9 ticks from 0 to 8000
y_ticks_pixel = np.linspace(pixel_extent[2], pixel_extent[3], 9)  # 9 ticks from 0 to 8000

x_ticks_degree = np.linspace(angular_extent[0], angular_extent[1], 9)
y_ticks_degree = np.linspace(angular_extent[2], angular_extent[3], 9)

fig, ax = plt.subplots(1, 1)

ax.set_xticks(x_ticks_pixel)
ax.set_yticks(y_ticks_pixel)

# Set tick labels
ax.set_xticklabels(x_ticks_degree)
ax.set_yticklabels(y_ticks_degree)

ax.imshow(convergence[0], vmin=-0.2, vmax=0.3)
ax.set_title("Convergence $(\Omega_m = 0.3,~\sigma_8=0.8)$, realization 0 / 648")
ax.set_xlabel("$x~(^\circ)$")
ax.set_ylabel("$y~(^\circ)$")

fig.savefig("plots/raytracing/convergence.pdf", bbox_inches="tight")
```
